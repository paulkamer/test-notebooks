```python
import cv2
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
```

```python
from ast import Pass


kernel_3x3 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
kernel_5x5 = np.array([[-1, -1, -1, -1, -1], [-1, 1, 2, 1, -1], [-1, 2, 4, 2, -1], [-1, 1, 2, 1, -1], [-1, -1, -1, -1, -1]])

img = cv2.imread("images/statue_of_liberty.jpeg", 0)

k3 = ndimage.convolve(img, kernel_3x3)
k5 = ndimage.convolve(img, kernel_5x5)
blurred = cv2.GaussianBlur(img, (17,17), 0)
g_hpf = img - blurred

f, axarr = plt.subplots(2, 2)
plt.title('Test')
axarr[0,0].imshow(cv2.cvtColor(k3, cv2.COLOR_BGR2RGB))
axarr[0,0].set_title('k3')

axarr[0,1].imshow(cv2.cvtColor(k5, cv2.COLOR_BGR2RGB))
axarr[0,1].set_title('k5')

axarr[1,0].imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
axarr[1,0].set_title('blurred')

axarr[1,1].imshow(cv2.cvtColor(g_hpf, cv2.COLOR_BGR2RGB))
axarr[1,1].set_title('g_hpf')

[axi.set_axis_off() for axi in axarr.ravel()]
plt.show()
None
```
